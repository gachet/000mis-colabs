
<h3 id="Definición-de-funciones">Definición de funciones<a class="anchor-link" href="#Definici%C3%B3n-de-funciones">¶</a></h3><p>Podemos definir funciones de dos formas:</p>
<ul>
<li>Usando una función <span style="color:green; font-family:Courier; font-weight: bold;">lambda</span>.</li>
<li>Usando <span style="color:green; font-family:Courier; font-weight: bold;">def</span>.</li>
</ul>


```python

f1 = lambda x: x ** 3
f2 = lambda x,y: x+y

def f3(x):
    if x > 2:
        return 0
    else:
        return 1
```


<p>Pero si definimos la función mediante "def"  no funciona con arrays numpy: yenemos que vectorizarla con la operación .vectorize</p>


```python

a = np.arange(4)
print(a)
print('f1(a) = ', f1(a[1]))
print(f3(a))

```

```python
vfunc = np.vectorize(f3)
print('f3(a) = ', vfunc(a))
```


<p><a id="vect"></a>
<a id="ze"></a>
Si hacemos operaciones en bloque para todos los elementos de un array numpy, se llama <strong>vectorización</strong> y tiene la ventaja de ser más rápido que utilizar un bucle que recorra todos los elementos y realice la misma operación. Veamos un ejemplo. </p>


```python

import time

z = np.linspace(-10,10,1000000)   # vector con un millón de elementos
zy = np.zeros_like(z)             # vector de ceros con la misma estructura que z

t = time.time()
for i in range(len(z)):           
    zy[i] = f1(z[i])
t1 = time.time()-t   
print('Sin vectorización: %.6f segundos' % t1)    


t = time.time()
zy = f1(z)
t1 = time.time()-t   
print('Con vectorización: %.6f segundos' % t1) 

```

<h2 id="Gráfica-de-una-función-de-una-variable">Gráfica de una función de una variable<a class="anchor-link" href="#Gráfica-de-una-función-de-una-variable">¶</a>



<p>La función <code>plot</code> del módulo <code>matplotlib.pyplot</code> dibuja la función aproximada. En realidad sustituye el dibujo de la función por  una poligonal de puntos de la función. Veamos un ejemplo:</p>
<p><a id="fe"></a>
<a id="lambda"></a>
Definimos la función a dibujar $f(x)=e^x$</p>



```python

f = lambda x: np.exp(x)

```


<p>Decidimos el intervalo $[a,b]$ en el que vamos a crear la función</p>


```python
##Decidimos el intervalo [a,b]
a = -1.; b = 1.
##Creamos una malla de 5 puntos equiespaciados en el intervalo [a,b] que se almacena en un array numpy
x = np.linspace(a,b,5)
print(x)
```

```python

##Obtenemos el array numpy correspondiente a las y

y = f(x)
print(y)
```


<p><a id="graf"></a>
Dibujamos la función</p>


```python

plt.plot(x,y)
plt.show()

```


<p>Con solo 5 puntos la aproximación es muy basta. Por defecto, <code>linspace</code> crea una malla de 50 puntos. Repitamos el proceso con 50 puntos.</p>


```python

x = np.linspace(a,b)
y = f(x)
plt.plot(x,y)
plt.show()

```


<p>Sigue siendo una poligonal, pero ahora el mallado es lo suficiente fino para que dé la impresión de ser una curva suave.</p>
<p>Añadamos algunos detalles al dibujo:</p>


```python

ox = 0*x                               # array numpy con 50 ceros. Para dibujar el eje OX
plt.plot(x,y, label = 'f')
plt.plot(x,ox,'k', label = 'Eje OX')   # k, de "black" para que nos dibuje el eje OX en negro
plt.title(u'Ejemplo dibujo función f') # u delante de las comillas porque hay un acento
plt.legend()                           # usando los "label", añade una leyenda
plt.show()

```

<h1 id="Otro Ejemplo">Otro Ejemplo<a class="anchor-link" href="#Otro Ejemplo">¶</a></h1>

```python

x = np.linspace(-1,2)              # malla
f = lambda x : x**3 - 2*x**2 + 1   # función 
OX = 0*x                           # eje OX
print("x es",x)
print("eje 0 OX es",OX)
```


<h3 id="Plot-de-una-función-real-de-dos-variables">Plot de una función real de dos variables<a class="anchor-link" href="#Plot-de-una-funci%C3%B3n-real-de-dos-variables">¶</a></h3>



<p>Creamos una malla para la coordenada $x$ y otra para la coordenada $y$. Es decir, por ejemplo, 20 puntos igualmente espaciados entre $-1$ y $1$ para cada coordenada.</p>


```python

xgrid = np.linspace(-1,1,20)
ygrid = np.linspace(-1,1,20)


```


<p>Creamos ahora dos matrices que contienen la malla bidimensional</p>


```python

X, Y = np.meshgrid(xgrid,ygrid)

```


<p>Definimos la función de dos variables</p>


```python

g = lambda x,y: x**2 + y**2

```


<p>Y representamos la función usando diferentes colores para cada intervalo de valores de la función</p>


```python

plt.contourf(X,Y,g(X,Y), cmap='jet')
plt.colorbar()
plt.show()

```


<p>También podemos representarla como una superficie</p>


```python

from mpl_toolkits.mplot3d import Axes3D

fig1  = plt.figure(figsize=(10,5))
ax1   = fig1.gca(projection='3d')
ax1.plot_surface(X, Y, g(X,Y), cmap='jet')
plt.show()

```


<p>O como una malla</p>


```python

fig2  = plt.figure(figsize=(10,5))
ax2   = fig2.gca(projection='3d')
ax2.plot_wireframe(X, Y, g(X,Y))
plt.show()

```
