# Interpolación Polinómica, Matriz de Vandermonde

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
np.set_printoptions(precision = 2)   # solo dos decimales
np.set_printoptions(suppress = True) # no usar notación exponencial
```

### Encontrar un polinomio de segundo grado que pase por los puntos (-2,-2); (-1, 1); (2,-1).

```python
x = np.array([-2,-1,2])
y = np.array([-2,1,-1])
N = len(y)
## V se puede calcular con
## np.vander(x,N) o bien
V = np.column_stack([x**(N-1-i)\
    for i in range(0,N)])
print('Matriz Vandermonde\n',V)
a = np.dot(np.linalg.inv(V),y)
print('coeficientes son ' ,a )
```

```python
res = lambda x: a[0]*x**2 + a[1]*x +a[2]
plt.figure(figsize = (10,6) )
plt.grid(True)
# plot data as scatter plot,
plt.scatter(x,y,s=120,c='green',\
        label='datos',zorder=2)
# plot best fit
xnew = np.linspace(-3,3,200)
plt.plot(xnew,res(xnew),c='red',\
linewidth=3,label='interpol',zorder=4)
plt.xlabel('$x$',fontsize=24)
plt.ylabel('$y$',fontsize=24)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc='upper right',\
fontsize=20)
plt.tight_layout()
plt.show()

```

### Encontrar un polinomio de cuarto grado que pase por los puntos (2,2), (3, 6), (4,5),(5,5),(6,6)

```python
x = np.array([2.,3.,4.,5.,6.])
y = np.array([2.,6.,5.,5.,6.])
N = len(x)
## V se puede calcular con
V = np.vander(x,N)
print('Matriz Vandermonde\n',V)
a = np.dot(np.linalg.inv(V),y)
print('coeficientes son ' ,a )
```

```python
res = lambda x: a[0]*x**4 + a[1]*x**3 +a[2]*x**2+a[3]*x+a[4]
plt.figure(figsize = (10,6) )
plt.grid(True)
# plot data as scatter plot,
plt.scatter(x,y,s=120,c='green',\
        label='datos',zorder=2)
# plot best fit
xnew = np.linspace(2,6,200)
plt.plot(xnew,res(xnew),c='red',\
linewidth=3,label='interpol',zorder=4)
plt.xlabel('$x$',fontsize=24)
plt.ylabel('$y$',fontsize=24)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc='lower right',\
fontsize=20)
plt.tight_layout()
plt.show()
```
